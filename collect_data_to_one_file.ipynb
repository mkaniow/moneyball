This script was created to prepare datasets to insert to database

```python
import pandas as pd
import numpy as np
import os
```

Data scraped from website

```python
details = 'nba_game_details'
games = 'nba_games'

files = os.listdir()

list_of_labels = ['GAME_ID', 'PLAYER', 'MIN', 'FGM', 'FGA', 'FG%', '3PM', '3PA', '3P%', 'FTM', 'FTA', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TO', 'PF', 'PTS', '+/-', 'TEAM']

list_of_labels_2 = ['GAME_ID', 'HOME', 'AWAY', 'PTS_HOME', 'PTS_AWAY', 'DATE']

game_details = pd.DataFrame(columns=list_of_labels)
nba_games = pd.DataFrame(columns=list_of_labels_2)

for item in files:
    if item.startswith(games):
        holder_1 = pd.read_csv(item)
        nba_games = pd.concat([nba_games, holder_1], ignore_index=True, sort=False)
        os.remove(item)
    elif item.startswith(details):
        holder_2 = pd.read_csv(item)
        game_details = pd.concat([game_details, holder_2], ignore_index=True, sort=False)
        os.remove(item)
    else:
        pass

nba_games.to_csv('combined_games_files.csv')
game_details.to_csv('combined_games_details_files.csv')
```

Data downoloaded from kaggle

```python
path = os.getcwd()

files = os.listdir(path + '/dane_nba_com')

game_details_2 = pd.DataFrame(columns=list_of_labels)
nba_games_2 = pd.DataFrame(columns=list_of_labels_2)

for item in files:
    if item.startswith(games):
        holder_1 = pd.read_csv(path + '/dane_nba_com/' + item)
        nba_games_2 = pd.concat([nba_games_2, holder_1], ignore_index=True, sort=False)
        os.remove(path + '/dane_nba_com/' + item)
    elif item.startswith(details):
        holder_2 = pd.read_csv(path + '/dane_nba_com/' + item)
        game_details_2 = pd.concat([game_details_2, holder_2], ignore_index=True, sort=False)
        os.remove(path + '/dane_nba_com/' + item)
    else:
        pass

nba_games_2.to_csv('combined_games_files_2.csv')
game_details_2.to_csv('combined_games_details_files_2.csv')
```

Combine csv files

```python
nba_games_combined = pd.concat([nba_games, nba_games_2])
game_details_combined = pd.concat([game_details, game_details_2])

nba_games_combined.to_csv('nba_games_combined.csv')
game_details_combined.to_csv('game_details_combined.csv')
```

Separate min and sec from details

```python
df_12 = pd.read_csv('games_details_all.csv', index_col=0)

new = df_12["MIN"]

mins = []
secs = []

for item in new:
    try:
        holder = item.split(':')
        holder[0] = int(holder[0])
        holder[1] = int(holder[1])
    except:
        holder = [np.NAN, np.NAN]
    mins.append(holder[0])
    secs.append(holder[1])

df_12['MINUTE'] = mins
df_12['SECOND'] = secs
df_12 = df_12.drop('MIN', axis=1)

df_12 = df_12.loc[:,['GAME_ID','PLAYER','MINUTE','SECOND','FGM','FGA','FG%','3PM','3PA','3P%','FTM','FTA','FT%','OREB','DREB','REB','AST','STL','BLK','TO','PF','PTS','+/-','TEAM']]

df_12.to_csv('games_details_all.csv')
```
